```python
from pathlib import Path
from zipfile import ZipFile

import numpy as np
import pandas as pd
from sklearn import decomposition
from sklearn.manifold import TSNE

DATA_DIR = Path().resolve().parent.joinpath('data')
```

```python
with ZipFile(DATA_DIR.joinpath('external').joinpath('07_online_news_popularity_shares.zip')).open('OnlineNewsPopularity/OnlineNewsPopularity.csv') as f:
    df = pd.read_csv(f, index_col=0)
df.describe()
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
pca = decomposition.PCA()
pca.fit(X)
```

```python
for column, score in zip(df.columns, pca.explained_variance_ratio_):
    print(f"{column}: {score:,.2f}")
```

```python
X_embedded = TSNE(
    n_components=3,
    learning_rate='auto',
    init='random',
    perplexity=3
).fit_transform(X)
X_embedded.shape
```
