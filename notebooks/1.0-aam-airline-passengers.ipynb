```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
from pandas.plotting import autocorrelation_plot, lag_plot

DATA_DIR = Path().resolve().parent.joinpath('data')
```

```python
kwargs = {
    'filepath_or_buffer': DATA_DIR.joinpath('airline_passengers.csv'),
    'index_col': 0,
    'parse_dates': [0],
}
```

```python
df = pd.read_csv(**kwargs)
```

```python
# =============================================================================
# Data Signature
# =============================================================================
df.info()
print(df.describe())
```

```python
# =============================================================================
# Plotting
# =============================================================================
df.plot(grid=True)

plt.figure(0)
df.pipe(autocorrelation_plot)
plt.show()

plt.figure(1)
df.pipe(lag_plot)
plt.grid()
plt.show()
```
