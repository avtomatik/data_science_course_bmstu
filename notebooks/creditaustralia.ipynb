```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv('../data/external/Credit_Card_Applications.csv', index_col=0)
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
X
```

```python
y
```

```python
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X, y)
```

```python
neigh.predict([[1.1]])
neigh.predict_proba([[0.9]])
```
